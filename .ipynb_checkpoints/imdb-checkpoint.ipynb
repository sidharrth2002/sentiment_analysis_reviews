```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
reviews = pd.read_csv('IMDB Dataset.csv')
reviews.head()
```

```python
reviews['sentiment_bool'] = reviews['sentiment'].apply(lambda x: 1 if x=='positive' else 0)
reviews.drop(columns=['sentiment'], inplace=True)
```

```python
reviews.head()
```

```python
reviews['review'].iloc[0]
```

## Preprocessing

```python
import re
def preprocessor(entry):
    #remove html tags
    entry = re.sub(r'<[^>]*>', '', entry)
    #remove emojis
    entry = re.sub(r'(?::|;|=)(?:-)?(?:\)|\(|D|P)', ' ', entry)
    return entry
```

```python
reviews['review'] = reviews['review'].apply(preprocessor)
```

## Word Clouds

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
stopwords = ENGLISH_STOP_WORDS.union(['movie', 'movies', 'film', 'character', 'just'])
```

```python
positive_reviews = reviews[reviews['sentiment_bool'] == 1]['review']
positives = WordCloud(stopwords = stopwords, background_color="white").generate(' '.join(positive_reviews))
plt.imshow(positives, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
negative_reviews = reviews[reviews['sentiment_bool'] == 0]['review']
positives = WordCloud(stopwords = stopwords, background_color="white").generate(' '.join(negative_reviews))
plt.imshow(positives, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Tokenization

```python
from nltk.tokenize import word_tokenize
word_tokens = [word_tokenize(review) for review in reviews['review']]
len_tokens = []

for i in range(len(word_tokens)):
    len_tokens.append(len(word_tokens[i]))
    
reviews['length'] = len_tokens
```

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
def tokenizer_stemmer(entry):
    return [porter.stem(word) for word in entry.split()]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words=stopwords, ngram_range=(1, 2), max_features=200, token_pattern=r'\b[^\d\W][^\d\W]+\b')
```

```python
X = tfidf.fit_transform(reviews.review)
y = reviews.sentiment_bool.values
```

```python
reviews_transformed = pd.DataFrame(X.toarray(), columns=tfidf.get_feature_names())
reviews_transformed.head()
```
