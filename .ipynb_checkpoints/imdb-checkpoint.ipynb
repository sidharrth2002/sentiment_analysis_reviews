```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
reviews = pd.read_csv('IMDB Dataset.csv')
reviews.head()
```

```python
reviews['sentiment_bool'] = reviews['sentiment'].apply(lambda x: 1 if x=='positive' else 0)
reviews.drop(columns=['sentiment'], inplace=True)
```

```python
reviews.head()
```

```python
reviews['review'].iloc[0]
```

## Preprocessing

```python
import re
def preprocessor(entry):
    #remove html tags
    entry = re.sub(r'<[^>]*>', '', entry)
    #remove emojis
    entry = re.sub(r'(?::|;|=)(?:-)?(?:\)|\(|D|P)', ' ', entry)
    return entry
```

```python
reviews['review'] = reviews['review'].apply(preprocessor)
```

## Word Clouds

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
stopwords = ENGLISH_STOP_WORDS.union(['movie', 'movies', 'film', 'character', 'just'])
```

```python
positive_reviews = reviews[reviews['sentiment_bool'] == 1]['review']
positives = WordCloud(stopwords = stopwords, background_color="white").generate(' '.join(positive_reviews))
plt.imshow(positives, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
negative_reviews = reviews[reviews['sentiment_bool'] == 0]['review']
positives = WordCloud(stopwords = stopwords, background_color="white").generate(' '.join(negative_reviews))
plt.imshow(positives, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from nltk.stem.porter import PorterStemmer
from nltk import sent_tokenize, word_tokenize
porter = PorterStemmer()
def tokenizer_stemmer(text):
    tokens = word_tokenize(text)
#     stems = []
    return [porter.stem(item) for item in tokens if item not in stopwords]
#     return [porter.stem(word) for word in entry.split()]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(ngram_range=(1, 2), max_features=200, token_pattern=r'\b[^\d\W][^\d\W]+\b')
```

```python
X = tfidf.fit_transform(reviews.review)
y = reviews.sentiment_bool.values
```

```python
reviews_transformed = pd.DataFrame(X.toarray(), columns=tfidf.get_feature_names())
reviews_transformed.head()
```

## Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
X = reviews_transformed
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
reg = LogisticRegression().fit(X_train, y_train)
y_predicted = reg.predict(X_test)
```

```python
print('Test Accuracy: ', accuracy_score(y_test, y_predicted))
conf_matrix = confusion_matrix(y_test, y_predicted)/len(y_test)
```

```python
import seaborn as sns
labels = ['True Neg','False Pos','False Neg','True Pos']
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(conf_matrix, annot=labels, fmt='')
```
